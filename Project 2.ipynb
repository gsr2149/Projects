## Introduction

 The Pythagorean theorem is named after Greek philosopher Pythagoras and it examines the relationship between sides $a, b, c$ of a right triangle. The Pythagorean theorem says that $a, b, c$ are lengths of sides of a right triange if and only if $a^2 + b^2 = c^2$. Here, $c$ represents the hypotenuse of the right triangle (longest side), while $a$ and $b$ are the other sides of the triangle. The Pythagorean triples are sets of three positive integers $(a, b, c)$ that satisy the Pythagorean theorem.
 
 An example of Pythagorean triples is $(3, 4, 5)$ since $3^2 + 4^2 = 5^2$. In order to find our triples, we can find Pythagorean tuples or positive integers $(a,b)$ such that $\sqrt(a^2 + b^2)$ is also an integer which equals $c$. In this report, we shall examine ways to find and define Pythagorean triples and analyze their structures.

## Defining Function `get_ptriples`

In part $1$, we want to create a function called `get_ptriples` that takes an arguement $n$ and returns a list of all Pythagorean triples $(a,b,c)$ where $1 \le a,b \le n$. From the `math` module, we call our `sqrt` function to aid us in defining our function. We begin by defining our `get_ptriples` function which essentially works as retreiver function that will append our criteria that is defined into a list. Using a nested `for` loop, the code will go through all the possible values of both $a$ AND $b$ from $1$ to $n$. 

Now, $c$ needs to be defined. Since we already established that Pythagorean triples are any value that $a^2 = b^2 = c^2$, then $c$ must be defined as $\sqrt(a^2 + b^2)$. Using an `if` statement, we can identify any `int` values of `c` and append them into our `ptriples = []` list. 

```python
from math import sqrt
```

```python
def get_ptriples(n):
    ptriples = [] #creates an empty list for us
    
    #Build a nested for loop
    for a in range(1,n+1):
        for b in range(1,n+1):
            c = sqrt(a**2 + b**2)
            if c == int(c): #Is c an integer? This checks it
                ptriples.append([a,b,int(c)])
    return ptriples
    
```

```python
get_ptriples(20) #Calls the first 20 Pythagorean triples
```

## Plotting Our Pythagorean Tuples

Our next step is to plot our Pythagorean Tuples or the values of $(a, b)$ that satisfy the criteria of our Pythagorean triples found above. Using list comprehension in Python, we can pick out our values of $a$ and $b$ that were already found. ## Analyzing the Structure of Pythagorean Tuples

In order to analyze the structure of pythagorean triples, one way is to look at the "primitive Pythagorean triples". What defines a primitive Pythagorean triple? Is is said that, $(a, b, c)$ is **primitive** if the greatest common divisor of $a, b$ and $c$ is $1$

We can calculate the greatest common divisor through the "Euclidean algorithm". The greatest common divisor is the largest number that divides two numbers without a remainder. According to csuohio.edu, The Euclidean algorithm "uses the principle that the GCD of a set of two numbers does not change if you replace the larger of the two with the remainder when you divide the larger of the two by the smaller." 

The algorithm works as follows:
- Start with integers $a$ and $b$, and assume that $a \le b$
- Subtract the smaller $a$ from the larger $b$, and replace the larger $b$ with the difference $b - a$
- Repeat step 2 until one of the numbers is $0$. The other will be the greatest common divisor

```python
def Euclidean_algorithm(a,b):

    
  # First, make sure a <= b
    if a > b: 
        a, b = b, a
    # Subtract as many multiples of a as we can 
    while True: 
        a, b = b % a, a
        
        # if one of the numbers is zero, stop
        if a == 0:
            break
    # The non-zero number is the GCD
    return b
```

Now that we have defined the `Euclidean_algorithm` function, we can define a function to find our primitive Pythagorean triples. The function defined below takes `ptriple` as an input. Then we define an `if` statement. From our `Euclidean_algorithm` function above, if $a$ (`ptriple[0]` or place 1), $b$ (`ptriple[1]` or place 2) == 1, then return True and this means the triple is primitive

```python
def is_primitive_triple(ptriple):
    if Euclidean_algorithm(ptriple[0], ptriple[1]) == 1:
        return True
    else:
        return False
```

```python
triples = get_ptriples(20) #Store get_ptriples function into variable 'triples'
```

Once more using list comprehension, we create a new list `prim_ptriple` that will contain only our primitive triples. Using the `is_primitive_triple` function, it will check each triple in `triples` and if `True` is returned, the triple can then be added to our list. 

```python
prim_ptriple = [ptriple for ptriple in triples if is_primitive_triple(ptriple)]
print(prim_ptriple) 
```

```python
#Now we want to continue the process of plotting our primitive tuples and draw out our a and b values
a_list = [a for (a,b,c) in prim_ptriple] 
b_list = [b for (a,b,c) in prim_ptriple]

print(a_list)
print(b_list)
```
